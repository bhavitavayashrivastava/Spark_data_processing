```python
in_text = spark.sparkContext.textFile('aitel_jio.txt')
```

```python
in_text.collect()
```

```python
text_flat = in_text.flatMap(lambda x: x.split(' '))
```

```python
text_flat.collect()
```

```python
text_map = text_flat.map(lambda x: (x,1))
```

```python
text_map.collect()
```

```python
text_rbk = text_map.reduceByKey(lambda x,y: x+y)
```

```python
text_rbk.collect()
```
