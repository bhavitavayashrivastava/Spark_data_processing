```python
import findspark
findspark.init()
from pyspark import SparkContext

sc = SparkContext()
```

```python
lines = sc.textFile("C:\\Users\\admin\\Desktop\\spark-data\\README_2020.txt")
```

```python
from pprint import pprint
words = lines.flatMap(lambda x: x.split(" "))
words.collect()
```

```python
wordcount = lines.map(lambda x: (x,1))
wordcount = wordcount.reduceByKey(lambda x,y: x+y)
print(wordcount.collect())
```
